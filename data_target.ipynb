```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "alle_kranten_volledig.csv"
df = pd.read_csv(file_path)

if 'relevance_window_in_hours' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df['relevance_window_in_hours'], bins=30, kde=True, color="orange", edgecolor="black")

    plt.xlabel("Relevance Window in Hours", fontsize=12)
    plt.ylabel("Frequentie", fontsize=12)
    plt.title("Verdeling van de Relevance Window in Hours", fontsize=14)

    plt.show()
else:
    print("De kolom 'relevance_window_in_hours' is niet gevonden in de dataset.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "alle_kranten_volledig.csv"
df = pd.read_csv(file_path)

if 'relevance_window_in_hours' in df.columns:
    # Log-transformatie toepassen om de verdeling beter te visualiseren
    df['log_relevance_window'] = np.log1p(df['relevance_window_in_hours'])

    # Histogram met log-transformatie
    plt.figure(figsize=(10, 5))
    sns.histplot(df['log_relevance_window'], bins=30, kde=True, color="blue", edgecolor="black")
    plt.xlabel("Log(Relevance Window in Hours)", fontsize=12)
    plt.ylabel("Frequentie", fontsize=12)
    plt.title("Log-getransformeerde Verdeling van de Relevance Window in Hours", fontsize=14)

    plt.show()
else:
    print("De kolom 'relevance_window_in_hours' is niet gevonden in de dataset.")
```

```python
print(df['relevance_window_in_hours'].describe())
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Naef model: altijd het gemiddelde van de target voorspellen
y_mean_pred = df['relevance_window_in_hours'].mean()
rmse_baseline = mean_squared_error(df['relevance_window_in_hours'], [y_mean_pred] * len(df), squared=False)
r2_baseline = r2_score(df['relevance_window_in_hours'], [y_mean_pred] * len(df))

print(f"Baseline R-score: {r2_baseline}")

print(f"Baseline RMSE (gemiddelde als voorspelling): {rmse_baseline}")
```

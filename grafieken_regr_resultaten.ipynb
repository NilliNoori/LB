```python
import matplotlib.pyplot as plt
import numpy as np

# Data voor de modellen
models = ["Baseline", "Neuraal Netwerk", "XGBoost"]

# RMSE en R²-score waarden
rmse_values = [55.97, 45.85, 45.19]
r2_values = [0.0, 0.28, 0.30]

# RMSE Grafiek zonder titel bovenin
fig1, ax1 = plt.subplots(figsize=(8, 5))
bars1 = ax1.bar(models, rmse_values, color="steelblue")

# Waarden boven de balken weergeven
for bar in bars1:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval + 1, round(yval, 2), ha='center', fontsize=12)

# Labels instellen
ax1.set_xlabel("Model")
ax1.set_ylabel("RMSE")

# Achtergrond doorzichtig maken
fig1.patch.set_alpha(0)
ax1.set_facecolor("none")

# Extra stijlingselementen toevoegen
plt.text(-0.5, -10, "RMSE", fontsize=16, fontweight="bold", color="navy", backgroundcolor="lightseagreen")

# Grafiek tonen
plt.show()

# R²-score Grafiek zonder titel bovenin
fig2, ax2 = plt.subplots(figsize=(8, 5))
bars2 = ax2.bar(models, r2_values, color="lightseagreen")

# Waarden boven de balken weergeven
for bar in bars2:
    yval = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2, yval + 0.02, round(yval, 2), ha='center', fontsize=12)

# Labels instellen
ax2.set_xlabel("Model")
ax2.set_ylabel("R²-score")

# Achtergrond doorzichtig maken
fig2.patch.set_alpha(0)
ax2.set_facecolor("none")

# Extra stijlingselementen toevoegen
plt.text(-0.5, -0.1, "R²-score", fontsize=16, fontweight="bold", color="navy", backgroundcolor="lightseagreen")

# Grafiek tonen
plt.show()
```
